```python
#Find the product_category with the highest average sales amount per transaction, 
#but only consider transactions where more than 3 units were sold.

avg_sales=sales.groupby('product_category')['sales_amount'].mean().reset_index(name='avg')
```

```python
#Identify the region where customers have the highest total sales on average, 
#but only include customers who have made purchases in more than one region.

customers=sales.groupby('customer_id')['region'].count().reset_index(name='counts')
```
