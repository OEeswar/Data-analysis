```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
laptop=pd.read_csv('/home/eeswar/Downloads/datasets/laptop.csv',sep='\t')
laptop
```

```python
sns.lineplot(x='year',y='qty',data=laptop,hue='model',style='model',size='model')
plt.show()
```

```python
sns.relplot(x='year',y='qty',data=laptop,col='model',kind='line',hue='model')
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='tip',y='total_bill',data=tips,kind='scatter',hue='time',col='day',style='smoker')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,fill=True,hue='time')
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.countplot(x='day',data=tips,hue='time')
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box',col='smoker')
plt.show()
```
